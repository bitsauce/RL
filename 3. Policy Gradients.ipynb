```python
import gym
import numpy as np
```

```python
env = gym.make("CartPole-v0")
action_space = env.action_space.n
state_space = env.observation_space.shape
print("Action space:", action_space)
print("State space:", state_space)
```

```python
import keras
from keras.models import Model
from keras.layers import Input, Dense
import keras.backend as K
```

```python
input_state = Input(shape=state_space)
x = Dense(512, activation="elu")(input_state)
x = Dense(64, activation="elu")(x)
action_prob = Dense(action_space, activation="softmax")(x)

def loss(rewards_and_baselines, action_prob):
    R_t = rewards_and_baselines[:, 0]
    b   = rewards_and_baselines[:, 1]
    
    # R_t:         [horizon]
    # b:           [horizon]
    # action_prob: [horizon, action_space]
    return K.sum((R_t - b) * K.log(action_prob))

model = Model(input_state, action_prob)
model.compile(optimizer="SGD", loss=loss)
```
